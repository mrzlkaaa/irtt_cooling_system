```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("123.csv", index_col="Timestamp")
df
```

```python
time_periods = [("2022-10-17", "2022-10-21"), ("2022-11-08","2022-11-11"), ("2022-11-15","2022-11-18"),
           ("2022-11-22","2022-11-25"), ("2022-11-29","2022-12-02"), ("2022-12-05","2022-12-09"), 
           ("2022-12-12","2022-12-16"), ("2022-12-19","2022-12-23")]
```

```python
data_periods = []
data_periods_mean = []
for period in time_periods:
    s,f = period
    data_periods.append(df.loc[s:f].dropna())
    data_periods_mean.append(df.loc[s:f].dropna().mean())
```

```python
for i in data_periods:
    plt.plot(np.arange(20, len(i), 1), i["481"][20:])
# plt.xticks(rotation=90)
```

```python
print(data_periods_mean)
```
