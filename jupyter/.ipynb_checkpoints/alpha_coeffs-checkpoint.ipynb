```python
X_interp = [1,2,4,5]
y_interp = [4.65, 4.855, 3.852, 3.675]
y3 = np.interp(3, X_interp, y_interp)
```

```python
y_interp = [4.65, 4.855, y3, 3.852, 3.675]
plt.bar(T2aHE_means.index, y_interp)
```

```python
T2aHE_mean = df.loc[:, ["T2aHE"]].mean(axis=0)
T2aHE_mean.values
t_coefs = T2aHE_mean.values/np.array(y_interp)
```

```python
t_coefs = df["Q2"].mean(axis=0)/5*t_coefs
```
